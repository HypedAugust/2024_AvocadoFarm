```python
 # 1.    

#  (sklearn)    .
import sklearn  #  
from sklearn.datasets import make_circles  #     

# 1000    .
n_samples = 1000

#    .
# make_circles       .
# noise    . (,       )
# random_state    . (   )
X, y = make_circles(n_samples, noise=0.03, random_state=42)

#   5    .
print(f"First 5 samples of X:\n {X[:5]}")  # X   5  
print(f"First 5 samples of y:\n {y[:5]}")  # y   5 () 


# 2.  DataFrame 

# pandas  DataFrame     .
import pandas as pd

# X 2   . X1   , X2    .
# y    (0  1)  .
circles = pd.DataFrame({
    "X1": X[:, 0],  # X    (     X1  )
    "X2": X[:, 1],  # X    (     X2  )
    "label": y  # y    () label  
})

# DataFrame  10   .
circles.head(10)  # head()      .


# 3.  (  )

#    matplotlib   .
import matplotlib.pyplot as plt

#     .
# plt.scatter  ( )   .
plt.scatter(
    x=X[:, 0],  # x X    (, X1 ) .
    y=X[:, 1],  # y X    (, X2 ) .
    c=y,  #    y ()   . ( 0, 1   )
    cmap=plt.cm.RdYlBu  # RdYlBu  , (Red), (Yellow), (Blue) .
)

# ;         .


# 4.    (  )

# X y   .
# X  (1000, 2),  1000    2   .
# y  (1000,),  1000      .
X.shape, y.shape


# 5. X y     

# X       y  .
X_sample = X[0]  # X    
y_sample = y[0]  # y     

#       y  .
print(f"Values for one sample of X: {X_sample} and the same for y: {y_sample}")

#    () .
print(f"Shapes for one sample of X: {X_sample.shape} and the same for y: {y_sample.shape}")
```

```python
# 1.2 Turn data into tensors and create train and test splits

import torch 
torch.__version__ #   .

type(X), X.dtype  # X   dtype( ) .

# Turn data into tensors
# (X, y)  .
# (tensor)       .
X = torch.from_numpy(X).type(torch.float) # X  NumPy   ,   float .
y = torch.from_numpy(y).type(torch.float) # y    , float  .

#  X y  5  .
X[:5], y[:5]  # X y  5  .


#  (train) (test)    .
from sklearn.model_selection import train_test_split

# train_test_split       .
# X_train, X_test     ,
# y_train, y_test      .
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y,
                                                    test_size=0.2, # 0.2 = 20% of data will be test & 80% will be train
                                                    random_state=42)
     
```

```python
# 2 Building a model 
# 2.1 setup device agnostic code so our code will run on an accelerator if there is one 

# construct a model by subclassing nn.module 
# designe a loss function and optimizer 
# create a training and test loop 

# Import PyTorch and nn
import torch
from torch import nn  # nn   nn.Module, nn.Linear    .

#  
device = 'cuda' if torch.cuda.is_available() else 'cpu'  
#  GPU    'cuda' ,   'cpu'  .
#         .


# subclasses nn.module 
# create nn.liear layers that are capable of handing the shapes of our data 
# defines a forward method that outlines the forward pass 
# instantiate an instance of our model class and send it to the target device 

#   (nn.Module    )
class CircleModelV0(nn.Module):  
    # nn.Module  CircleModelV0  .
    
    def __init__(self):  #   .
        super().__init__()  #  (nn.Module) __init__      .
        
        #     (Linear layer) .
        #   (features)  2,    5  .
        # :      2  5  .
        self.layer_1 = nn.Linear(in_features=2, out_features=5)  
        
        #     .
        #   5    1    (    ).
        self.layer_2 = nn.Linear(in_features=5, out_features=1)  
        
    def forward(self, x):  # forward       .
        #   x layer_1   , layer_2    .
        return self.layer_2(self.layer_1(x))  
        # layer_1    layer_2    .

#      
model_0 = CircleModelV0().to(device)  
# CircleModelV0  ,  'device'  (GPU  CPU) .

#  nn.Sequential()  (   )
model_0 = nn.Sequential(  
    nn.Linear(in_features=2, out_features=5),  #   : 2  5  
    nn.Linear(in_features=5, out_features=1)   #   : 5  1   ()
).to(device)  
# nn.Sequential()          .
#     CircleModelV0      .
#    'device'  .

#  
#         .         .
#     ,        .
#   torch.inference_mode()        . 
#            .
with torch.inference_mode():  
    # inference_mode()     ,    (   ).
    untrained_preds = model_0(X_test.to(device))  
    # X_test  'device'  ,   .
    #       (    ).

print(f"First 10 predictions:\n{torch.round(untrained_preds[:10])}")  #  10   .
print(f"First 10 labels:\n{y_test[:10]}")  #   10   .
```

```python
# 2.1 Setup loss function and optimizer 
# For regression you might want MAE or MSE (mean absolute error or mean squared error).
# For classification you might want binary cross entropy or categorical cross entropy (cross entropy).

loss_fn = nn.BCEWithLogitsLoss() 
# BCEWithLogitsLoss:         . 
#  0  1      .
optimizer = torch.optim.SGD(params=model_0.parameters(),
                            lr=0.1) #     . 

# Calculate accuracy / testing 100 exmaples 
#   (weights)      .
def accuracy_fn(y_true, y_pred):
    correct = torch.eq(y_true, y_pred).sum().item()
    # torch.eq() (y_true) (y_pred)   .  True(1),  False(0) .
    # sum():         .
    # item():      .
    acc = (correct/len(y_pred))*100
    #   .        , (%) .
    return acc #   .


```

```python
# 3 Train model 
# training loop with the following steps: 
# forward pass/ calculate the loss / optimizer zero grad / loss backward / optimizer step  

# 3.1 going from raw logits -> prediction probabilities -> prediction labels. 
# model outputs = raw logits 
# convert logits -> prediction probabilities by passing them to activation function 
# convert prediction probabilities to prediction labels by taking the argmax()

# View the first 5 outputs of the forward pass on the test data
model_0.eval()  
#    .      .

with torch.inference_mode():  
    #  **torch.inference_mode()** forward pass . 
    #     .     .
    y_logits = model_0(X_test.to(device))[:5]  
    #  X_test     ,    5   .
    #     'logits',    .
y_logits  
#    5 logit  .
#  forward pass       .
#    forward pass ,    logits  . 
# forward pass     .

# use the sigmoid activation function on our model logits to turn them into prediction probabilities
y_pred_probs = torch.sigmoid(y_logits)  
# logit  sigmoid     .
# sigmoid  logit  0 1     .
y_pred_probs  
# sigmoid     .

# find the prediction labels
y_preds = torch.round(y_pred_probs)  
#      .
#  0.5  1, 0.5  0     .

# in full
y_pred_labels = torch.round(torch.sigmoid(model_0(X_test.to(device))[:5]))  
#      .
#  logit  sigmoid   ,      .
#    5 X_test    .

# Check for equality
print(torch.eq(y_preds.squeeze(), y_pred_labels.squeeze()))  
# y_preds y_pred_labels  .
# squeeze()         ,  .
#      y_pred_labels   y_preds    .

# Get rid of extra dimension
y_preds.squeeze()  
# squeeze()    1  .
#  , (5, 1)   (5,)      .
#     ,     .
```

```python
# 3.2 building a training and testing loop 

torch.manual_seed(42)
torch.cuda.manual_seed(42)
#          . 
#    . `42`   ,         .

# set the number of epochs 
epochs = 100  
#           . 
#  100  .

# put data to target device 
X_train, y_train = X_train.to(device), y_train.to(device) 
X_test, y_test = X_test.to(device), y_test.to(device)
#  (X_train, y_train)  (X_test, y_test) 
#   (: CPU  GPU)  .

# build training and evaluation loop 
for epoch in range(epochs):  
    #       .    100 .

    ### training 
    model_0.train()  
    #    .     , 
    # Dropout       .

    # 1. forward pass 
    y_logits = model_0(X_train).squeeze()  
    #       . 
    # y_logits raw logits(    ). `.squeeze()`  1  .
    
    y_pred = torch.round(torch.sigmoid(y_logits))  
    # logits        ,   0  1 . 
    # , logits ->   ->  (0  1)  .

    # 2. calculate loss/accuracy 
    # loss = loss_fn(torch.sigmoid(y_logits)), # nn.BCELoss expects prediction probabilities as input 
    loss = loss_fn(y_logits, y_train)  
    #  .  `BCEWithLogitsLoss`   
    # raw logits     .
    
    acc = accuracy_fn(y_true=y_train, y_pred=y_pred)  
    #  . `y_train`( ) `y_pred`( )   .

    # 3. Optimizer zero grad 
    optimizer.zero_grad()  
    #    (gradients) . 
    #               .

    # 4. Loss backward (backpropagation)
    loss.backward()  
    # (backpropagation)     . 
    #      ,      .

    # 5. Optimizer step (Gradient Descent)
    optimizer.step()  
    #     ()  
    #     .

    ### testing 
    model_0.eval()  
    #    .     , 
    #     (Dropout ) .

    with torch.inference_mode():  
        # inference_mode()         , 
        #    .
        
        # 1. forward pass 
        test_logits = model_0(X_test).squeeze()  
        #      .
        
        test_pred = torch.round(torch.sigmoid(test_logits))  
        # logits   ,   0  1 .

        # 2. calculate test loss/acc 
        test_loss = loss_fn(test_logits, y_test)  
        #     .
        
        test_acc = accuracy_fn(y_true=y_test, y_pred=test_pred)  
        #     .

# Print out what's happenin'  
#   ( 10 )        . 
# (Loss) (Accuracy)       .

if epoch % 10 == 0:  
    print(f"Epoch: {epoch} | Loss: {loss:.5f}, Acc: {acc:.2f}% | Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")  
    # 10    , ,    ,  . 
    # `{loss:.5f}`  5  .
```

```python
# improving a model (from a model perspective)

# Options are all values we (as machine learning engineers and data scientists) 
# can change, they are referred to as hyperparameters. 
# :       . ,    .

X_train[:5], y_train[:5]  #       5  .

# Create a model 
#    ,   nn.Module  
class CircleModelV1(nn.Module):
    def __init__(self):
        super().__init__()  #  (nn.Module) __init__  
        #   : 2   10   (2 -> 10)
        self.layer_1 = nn.Linear(in_features=2, out_features=10)
        #   : 10   10   (10 -> 10)
        self.layer_2 = nn.Linear(in_features=10, out_features=10)
        #   : 10   1   (10 -> 1,  )
        self.layer_3 = nn.Linear(in_features=10, out_features=1)
    
    def forward(self, x):  #     (Forward Pass)
        # x  layer_1 -> layer_2 -> layer_3    .
        return self.layer_3(self.layer_2(self.layer_1(x)))

#     'device'(GPU  CPU) 
model_1 = CircleModelV1().to(device)

# Create a loss function 
#  (BCEWithLogitsLoss):     ,   
loss_fn = nn.BCEWithLogitsLoss()

# Create an optimizer 
# (SGD):       
# lr:  (learning rate),     
optimizer = torch.optim.SGD(params=model_1.parameters(),
                            lr=0.1)

# write a training and evaluation loop for model_1
#      

torch.manual_seed(42)  #     
torch.cuda.manual_seed(42)

# train for longer 
epochs = 1000  #  1000   (      )

# put data on the target device 
#      (GPU CPU)
X_train, y_train = X_train.to(device), y_train.to(device)
X_test, y_test = X_test.to(device), y_test.to(device)

#   
for epoch in range(epochs):
    ### training (  )
    model_1.train()  #    
    
    # 1. forward pass 
    #       ()
    y_logits = model_1(X_train).squeeze()  # .squeeze()   
    y_pred = torch.round(torch.sigmoid(y_logits))  #         

    # 2. calculate the loss/acc
    #  :      
    loss = loss_fn(y_logits, y_train)
    #  :     
    acc = accuracy_fn(y_true=y_train,
                      y_pred=y_pred)
    
    # 3. optimizer zero grad 
    #      ( )
    optimizer.zero_grad()

    # 4. loss backward (backpropagation)
    #      
    loss.backward()

    # 5. optimizer step (gradient descent)
    #      ( )
    optimizer.step()

    ### testing (  )
    model_1.eval()  #     (  )
    with torch.inference_mode():  #    ( )
        # 1. forward pass 
        test_logits = model_1(X_test).squeeze()  #     
        test_pred = torch.round(torch.sigmoid(test_logits))  #      

        # 2. calculate loss
        #   :       
        test_loss = loss_fn(test_logits, y_test)
        #   
        test_acc = accuracy_fn(y_true=y_test, y_pred=test_pred)

    # print out what is happening (    )
    if epoch % 100 == 0:  # 100   
        print(f"Epoch: {epoch} | Loss: {loss:.5f}, Acc: {acc:.2f}% | Test loss: {test_loss:.5f}, Test acc: {test_acc:.2f}%")
        #       
```

```python
# 5.1 Preparing data to see if our model can fit a straight line
# One way to troubleshoot to a larger problem is to test out a smaller problem.

weight = 0.7  #   
bias = 0.3    #   
start = 0     #  
end = 1       #  
step = 0.01   #     

#   (X  0 1 0.01  , 1  )
X_regression = torch.arange(start, end, step).unsqueeze(dim = 1)

#    (y = wx + b)  y 
y_regression = weight * X_regression + bias

#   ,  5   
print(len(X_regression))
X_regression[:5], y_regression[:5]

# create train and test splits 
train_split = int(0.8 * len(X_regression))
X_train_regression, y_train_regression = X_regression[:train_split], y_regression[:train_split]
X_test_regression, y_test_regression = X_regression[train_split:], y_regression[train_split:]

# check the length of each 
len(X_train_regression), len(X_test_regression), len(y_train_regression), len(y_test_regression)

plot_predictions(train_data=X_train_regression,
                 train_labels=y_train_regression,
                 test_data=X_test_regression,
                 test_labels=y_test_regression)
```

```python
# 5.2 Adjusting model_1 to fit a straight line 

# same architecture as model_1 (but using nn.Sequential())
# nn.Sequential     .
# Sequential      .
model_2 = nn.Sequential(
    nn.Linear(in_features=1, out_features=10),  #   :  1,  10  (  )
    nn.Linear(in_features=10, out_features=10),  #   : 10   10   ( )
    nn.Linear(in_features=10, out_features=1)  #   : 10   1   (  )
).to(device)  #  GPU CPU  (device   ).

model_2  #   .

# loss and optimizer
#  (Loss Function) (Optimizer) .
loss_fn = nn.L1Loss()  # L1  :     (   )
optimizer = torch.optim.SGD(params=model_2.parameters(), lr=0.01)  
# SGD  .    , (lr) 0.01 .

# train the model 
#    
torch.manual_seed(42)  #   :       
torch.cuda.manual_seed(42)  # CUDA(GPU)   

# set the number of epochs
epochs = 1000  #    . (1000  )

# Put the data on the target device
#   (CPU  GPU) 
X_train_regression, y_train_regression = X_train_regression.to(device), y_train_regression.to(device)
X_test_regression, y_test_regression = X_test_regression.to(device), y_test_regression.to(device)

# training
#     .
for epoch in range(epochs):
    # 1. Forward pass:      .
    y_pred = model_2(X_train_regression)
    
    # 2.  :     (Loss) .
    loss = loss_fn(y_pred, y_train_regression)
    
    # 3. (Gradient) :    0 .
    optimizer.zero_grad()
    
    # 4. (Backpropagation):      .
    loss.backward()
    
    # 5.  :      .
    optimizer.step()

    # testing 
    #      .
    model_2.eval()  #     (  ).
    with torch.inference_mode():  #   ,  .
        test_pred = model_2(X_test_regression)  #     .

    # print out what is happening
    # (epoch) 100     .
    if epoch % 100 == 0:  
        # .5f   5    .
        print(f'Epoch:{epoch} | Loss : {loss:.5f} | Test loss: {test_loss:.5f}')


# Turn on evaluation mode
model_2.eval()

# Make predictions (inference)
with torch.inference_mode():
  y_preds = model_2(X_test_regression)

# Plot data and predictions
plot_predictions(train_data=X_train_regression.cpu(), 
                 train_labels=y_train_regression.cpu(),
                 test_data=X_test_regression.cpu(),
                 test_labels=y_test_regression.cpu(),
                 predictions=y_preds.cpu()); 
```
