```python
# dfs 매서드 정의 
def dfs(graph, v, visited):
    # 현재 노드 방문 처리 
    visited[v] = True
    print(v, end=' ')

    # 현재 노드와 연결된 다른 노드를 재귀적으로 방문 
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
    

# 각 노드가 연결된 정보를 표현 
graph = [
    [],           # 0번 노드는 사용되지 않음 (빈 리스트)
    [2, 3, 8],    # 1번 노드는 2번, 3번, 8번 노드와 연결
    [1, 7],       # 2번 노드는 1번, 7번 노드와 연결
    [1, 4, 5],    # 3번 노드는 1번, 4번, 5번 노드와 연결
    [3, 5],       # 4번 노드는 3번, 5번 노드와 연결
    [3, 4],       # 5번 노드는 3번, 4번 노드와 연결
    [7],          # 6번 노드는 7번 노드와 연결
    [2, 6, 8],    # 7번 노드는 2번, 6번, 8번 노드와 연결
    [1, 7]        # 8번 노드는 1번, 7번 노드와 연결
]

# 각 노드가 방문된 정보를 표현 
visited = [False]*9

# 정의된 DFS 함수 호출 
dfs(graph, 1, visited)
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# 그래프 정보 정의
graph = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]

# 그래프 생성
G = nx.Graph()

# 각 노드와 간선 추가
for node, connections in enumerate(graph):
    for connected_node in connections:
        G.add_edge(node, connected_node)

# 그래프 시각화
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title("Graph Visualization")
plt.show()
```
