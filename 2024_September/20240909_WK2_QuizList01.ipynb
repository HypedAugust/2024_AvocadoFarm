
스타상(*) 효과

무지의 다트 게임에 스타상(*)이 추가되었어요. 규칙은 다음과 같습니다:

다트를 3번 던집니다.
각 던지기의 점수는 0부터 10까지입니다.
점수 뒤에 'S', 'D', 'T' 중 하나가 옵니다.
'S', 'D', 'T' 뒤에 '*'이 올 수 있습니다.
'*'는 해당 점수와 바로 전 점수를 2배로 만듭니다.
'*'가 첫 번째 나오면 해당 점수만 2배가 됩니다.

입력 형식: "점수(S|D|T)[]점수(S|D|T)[]점수(S|D|T)[*]"로 이루어진 문자열
출력 형식: 3번의 던지기에서 얻은 점수의 합계
입출력 예제:
예제 | dartResult | answer | 설명
1    | "1S2S3S" | 24     | (12)2 + (22) + 3 = 4 + 4 + 3 = 11
2    | "1D2S3S"  | 19     | 1^2 + (2*2) + 3 = 1 + 4 + 3 = 8

아차상(#) 효과

무지의 다트 게임에 아차상(#)이 추가되었어요. 규칙은 다음과 같습니다:

다트를 3번 던집니다.
각 던지기의 점수는 0부터 10까지입니다.
점수 뒤에 'S', 'D', 'T' 중 하나가 옵니다.
'S', 'D', 'T' 뒤에 '#'이 올 수 있습니다.
'#'은 해당 점수를 마이너스로 만듭니다.

입력 형식: "점수(S|D|T)[#]점수(S|D|T)[#]점수(S|D|T)[#]"로 이루어진 문자열
출력 형식: 3번의 던지기에서 얻은 점수의 합계
입출력 예제:
예제 | dartResult | answer | 설명
1    | "1S2D#3T"  | 28     | 1 + (-2^2) + 3^3 = 1 + (-4) + 27 = 24
2    | "1D#2S10S" | 9      | (-1^2) + 2 + 10 = -1 + 2 + 10 = 11

스타상과 아차상 조합

무지의 최종 다트 게임은 스타상(*)과 아차상(#)을 모두 사용합니다. 규칙은 다음과 같습니다:

다트를 3번 던집니다.
각 던지기의 점수는 0부터 10까지입니다.
점수 뒤에 'S', 'D', 'T' 중 하나가 옵니다.
'S', 'D', 'T' 뒤에 '*'이나 '#'이 올 수 있습니다.
'*'는 해당 점수와 바로 전 점수를 2배로 만듭니다.
'#'은 해당 점수를 마이너스로 만듭니다.
'*'와 '#'의 효과는 중첩될 수 있습니다.

입력 형식: "점수(S|D|T)[|#]점수(S|D|T)[|#]점수(S|D|T)[*|#]"로 이루어진 문자열
출력 형식: 3번의 던지기에서 얻은 점수의 합계
입출력 예제:
예제 | dartResult | answer | 설명
1    | "1D#2S3S" | 5      | (-1^2)2 + (22) + 3 = -2 + 4 + 3 = 5
2    | "1S2T#3S" | 23     | (1*2) + (-2^3) + 3 = 2 + (-8) + 3 = -3
