````python
# ì „ì²´ ë¦¬ìŠ¤íŠ¸ë¥¼ ê°€ì§€ê³  ì™€ì•¼í•œë‹¤. 
# ë¦¬ìŠ¤íŠ¸ë¥¼ ì²´í¬ í•´ì•¼í•œë‹¤. 
# ì½ê³  ë‹«ê³ ë¥¼ forë¬¸ìœ¼ë¡œ ë°˜ë³µí•´ì•¼ í•œë‹¤. 
# ì €ë²ˆì—” í•œë²ˆì— íŒŒì¼ì„ ì¼ê³  ì´ë²ˆì—” í•œë²ˆì— íŒŒì¼ì„ ì½ëŠ”ë‹¤. 

import os

def read_files(file_path):
    # ê²°ê³¼ ë¦¬ìŠ¤íŠ¸
    outputs = []

    
    for file in os.listdir(file_path):
        
        if file.endswith(".txt"):
            
            target_path = os.path.join(file_path, file)
            
            with open(target_path, 'r') as f:
                outputs.append(f.read()) 

    return " ".join(outputs)

print(read_files(r"C:\Users\82106\Desktop\27-1"))
````
