```python
import numpy as np
import matplotlib.pyplot as plt 

arr = np.random.normal(0,1,1000)
plt.hist(arr,bins = 100)
plt.show()
```

```python
# np.random.rand()
arr = np.random.rand(1000)
plt.hist(arr, bins = 100)
plt.show()
```

```python
# np.random.randn() 정규분포, 가우시안 분포
arr = np.random.randn(1000)
plt.hist(arr, bins =100)
plt.show()
```

```python
# 랜덤한 정수만 출력하는 함수 
# np.random.randint()
arr = np.random.randint(low = 1, high= 5, size = (3,5)) # hyper parameter는 반드시 1개 이상 작성
print(arr)
plt.hist(arr,bins =100)
plt.show()
```

```python
# rand(), randn(), randint() 3가지가 가장 많이 쓰인다. 
```

```python
# seed 값을 통한 난수 생성 제어 

arr = np.random.rand(10)
print('난수 발생1 \n', arr)


arr = np.random.rand(10)
print('난수 발생2 \n', arr)
```

```python
np.random.seed(1)
arr = np.random.rand(10)
print('난수 발생 1\n', arr)

np.random.seed(1)
arr = np.random.rand(10)
print('난수 발생 2\n', arr)
```
