```python
def recursive_dfs(graph, vertex, visited=None):
    if visited is None:
        visited = set()
    visited.add(vertex)
    print(vertex, end=' ')  # 공백 추가해서 출력 형식 통일

    for neighbor in graph[vertex]:
        if neighbor not in visited:
            recursive_dfs(graph, neighbor, visited)
    return visited

# DFS 호출
recursive_dfs(graph, 'A')
```

```python
# 자료 구조
# stack (스택)

class stack:
    def __init__(self):
        # 스택을 구현하기 위한 리스트 초기화
        self.items = []  # 빈 리스트를 생성하여 스택 항목을 저장할 준비를 함

    def push(self, item):
        # 스택에 항목을 추가하는 함수
        self.items.append(item)  # 리스트의 끝에 항목을 추가하여 스택에 쌓음

    def pop(self):
        # 스택의 맨 위에 있는 항목을 제거하고 반환하는 함수
        return self.items.pop()  # 리스트의 끝에서 항목을 제거하고 반환함

    def peek(self):  # 스택 맨 위의 항목을 보여주는 함수
        # 스택의 맨 위 항목을 반환하되, 스택이 비어 있으면 None을 반환함
        return self.items[-1] if self.items else None  # 비어있을 때는 맨 뒤 항목이 없으므로 None을 반환하여 오류를 방지
    

# 스택 클래스 사용 예시 
my_stack = stack() # 스택 객체 생성 

# 스택에 항목 추가 push
my_stack.push(10)
my_stack.push(20)
my_stack.push(30)

# 현재 상태 확인 
print('맨위', my_stack.peek())
```

```python
# Singly Linked List의 간단한 기본 구현 코드

# 노드 클래스 정의
class Node:
    def __init__(self, value):
        self.item = value  # 노드에 저장된 값 (item)
        self.next = None  # 다음 노드를 가리키는 포인터 (next)

# LinkedList 클래스 정의
class LinkedList:
    def __init__(self):
        self.head = None  # 시작 지점 (첫 번째 노드) 지정

    # 노드 추가 예시
    def insert_at_beginning(self, value):
        new_node = Node(value)  # 새로운 Node 객체 생성 (LinkedList는 Node를 포함하는 형태로 사용)
        new_node.next = self.head  # 새 노드가 현재 헤드를 가리킴
        self.head = new_node  # 새로운 노드를 head로 지정

    # 리스트 끝에 새로운 노드 추가 (Insertion at End)
    def insert_at_end(self, value):
        new_node = Node(value)  # 새로운 노드 생성 (Node 객체가 생성됨)
        if self.head is None:  # 리스트가 비어 있는 경우 (아직 노드가 없음)
            self.head = new_node  # head를 새 노드로 지정 (새로운 노드가 리스트의 첫 번째 노드가 됨)
            return
        last_node = self.head
        while last_node.next:  # 마지막 노드를 찾기 위해 반복 (마지막 노드는 next가 None임)
            last_node = last_node.next
        last_node.next = new_node  # 마지막 노드의 next를 새 노드로 지정 (새로운 노드를 리스트 끝에 추가)

    # 리스트에서 특정 위치에 노드 삽입 (Insertion at Position)
    def insert_at_position(self, value, position):
        new_node = Node(value)  # 새로운 노드 생성 (Node 객체가 생성됨)
        if position == 0:  # 위치가 0인 경우 맨 앞에 삽입
            new_node.next = self.head  # 새 노드의 다음 노드를 현재 head로 지정
            self.head = new_node  # head를 새 노드로 업데이트
            return
        current = self.head
        pos = 0
        while pos < position - 1 and current is not None:  # 지정된 위치 전까지 이동 (현재 위치가 원하는 위치 전까지 반복)
            current = current.next
            pos += 1
        if current is None:
            print("Index out of range.")  # 위치가 리스트의 길이를 벗어나는 경우 (해당 위치에 삽입 불가)
        else:
            new_node.next = current.next  # 새 노드의 next를 현재 노드의 다음 노드로 지정
            current.next = new_node  # 현재 노드의 next를 새 노드로 지정 (새 노드를 현재 노드 뒤에 삽입)

    # 리스트에서 특정 위치의 노드 반환 (Retrieve an Item)
    def get(self, position):
        current = self.head
        pos = 0
        while current is not None:  # 리스트를 순회하면서 지정된 위치까지 이동
            if pos == position:
                return current.item  # 지정된 위치의 노드 값을 반환
            current = current.next
            pos += 1
        return None  # 위치가 리스트의 범위를 벗어나는 경우 None 반환

    # 리스트에서 특정 위치의 노드 삭제 (Deletion at Position)
    def delete_at_position(self, position):
        if self.head is None:  # 리스트가 비어 있는 경우 (삭제할 노드가 없음)
            print("List is empty.")
            return
        if position == 0:  # 첫 번째 노드를 삭제하는 경우
            self.head = self.head.next  # head를 두 번째 노드로 업데이트 (첫 번째 노드를 제거)
            return
        current = self.head
        pos = 0
        while pos < position - 1 and current.next is not None:  # 지정된 위치 전까지 이동
            current = current.next
            pos += 1
        if current.next is None:
            print("Index out of range.")  # 위치가 리스트의 길이를 벗어나는 경우 (해당 위치에 노드가 없음)
        else:
            current.next = current.next.next  # 현재 노드의 다음 노드를 건너뛰어 삭제 (지정된 위치의 노드 삭제)

    # 리스트의 모든 노드 출력 (Display All Nodes)
    def display(self):
        current = self.head
        while current:  # 리스트를 순회하면서 각 노드의 값을 출력
            print(current.item, end=" -> ")  # 각 노드의 값을 출력
            current = current.next
        print("None")  # 마지막에는 None 출력 (리스트의 끝을 나타냄)

# 사용 예시
linked_list = LinkedList()  # 빈 LinkedList 객체 생성
linked_list.insert_at_beginning(10)  # 리스트 맨 앞에 10 삽입
linked_list.insert_at_end(20)  # 리스트 끝에 20 삽입
linked_list.insert_at_end(30)  # 리스트 끝에 30 삽입
linked_list.insert_at_position(15, 1)  # 위치 1에 15 삽입
linked_list.display()  # 출력: 10 -> 15 -> 20 -> 30 -> None
print("Retrieved item at position 1:", linked_list.get(1))  # 위치 1의 노드 값 출력 (15)
linked_list.delete_at_position(1)  # 위치 1의 노드 삭제
linked_list.display()  # 출력: 10 -> 20 -> 30 -> None
```
