```python
#   while   .   . 

def recursive_function(i):
    # 100     
    if i == 100:
        return  
    print(i,' ', i+1, '  .')
    recursive_function(i+1)
    print(i,'  .')

recursive_function(1)
```

```python
# factorial  

def factorial_recursive(n):
    if n <= 1:
        return 1
    return n*factorial_recursive(n-1)

print(f' :{factorial_recursive(5)}')
```

```python
#   
def gcd(a,b):
    if a%b == 0:
        return b 
    else:
        return gcd(b, a%b)
        
```
