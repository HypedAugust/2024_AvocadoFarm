```python
# 클래스 선언
class Car(object):
    '''
    Car Class
    Author : Me
    Date : 2019.11.08
    Description : Class, Static, Instance Method
    '''

    # Class Variable
    price_per_raise_after = 1.2  # 클래스 변수 이름 통일
    
    
    # 이거는 인스턴스 변수야? 
    def __init__(self, company, details):
        self._company = company
        self._details = details
        
    def __str__(self):
        return 'str : {} - {}'.format(self._company, self._details)

    def __repr__(self):
        return 'repr : {} - {}'.format(self._company, self._details)

    # Instance Method1
    def detail_info(self):
        print('Current Id : {}'.format(id(self)))
        print('Car Detail Info : {} {}'.format(self._company, self._details.get('price')))
    
    # Instance Method2
    def get_price(self):
        return 'Before Car Price -> company : {}, price : {}'.format(self._company, self._details.get('price'))
    
    # Instance Method3 
    def get_price_cal(self): 
        return 'After Car Price -> company : {}, price : {}'.format(self._company, self._details.get('price') * Car.price_per_raise_after)
        
    # class method 클래스 변수를 핸들링 하려면 클래스 매소드를 써야 한다. 
    @classmethod
    def raise_price(cls, per):
        if per <= 1:
            print('please enter 1 or more')
            return 'Please enter a value greater than 1'
        cls.price_per_raise_after = per
        return 'Succeed! price increased.'
    
    # static method - 아무것도 전달 받지 않는다. 유연하게 사용할 수 있음. 
    @staticmethod
    def is_bmw(inst):
        if inst._company == 'BMW':
            return 'Yes {}'.format(inst._company)
        return 'not BMW'
        
# 자동차 인스턴스    
car1 = Car('Bmw', {'color' : 'Black', 'horsepower': 270, 'price': 5000})
car2 = Car('Audi', {'color' : 'Silver', 'horsepower': 300, 'price': 6000})

# 인스턴스 메소드 써보기 
car1.detail_info()

# 가격정보 (직접 접근)
print(car1._details.get('price'))

# 가격정보 (인상 전)
print(car1.get_price())
print(car2.get_price())

# 가격정보 (인상 후)
print(car1.get_price_cal())
print(car2.get_price_cal())

# 가격 인상 (클래스 메소드 미사용) 
Car.price_per_raise_after = 1.5
print(car1.get_price_cal())
print(car2.get_price_cal())

# 가격 인상 (클래스 메소드 사용)
print(Car.raise_price(1.6))

# 스태틱 테스트 - 상당히 유연하게 활용가능하다. 
print(car1.is_bmw(car1))
print(Car.is_bmw(car2))


# 정리 
# self 붙어 있으면 인스턴스 
# class는 다 공유함.

```
