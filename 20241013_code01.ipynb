```python
# 부서별로 신청한 금액이 들어있는 배열 d와 예산 budget이 매개변수
# 최대 몇 개의 부서에 물품을 지원할 수 있는지 return 
from itertools import permutations

def solution(d, budget):
    answer = 0
    final_answer = 0 
    for i in d: 
        while answer <= budget:
            answer += i
            new_d = list(permutations(d, i))
            for j in new_d:
                final_answer = len(max(j))
    return final_answer
```

```python
# BFS 문제임. 최단 경로는 보통 BFS, 큐를 써서 해결함. 
# 큐에 현재 위치, 현재까지 이동한 칸수를 저장하는게 좋음. 
# 동서남북 방향으로 한칸씩 이동하는것을 설정할 것 
# 방문 여부 기록할 것. 
# 상대방 진영에 도착 시 이동칸수를 반환할 것 

from collections import deque

def solution(maps):
    dx = [1,-1,0,0]
    dy = [0,0,1,-1]

    n,m = len(maps), len(maps[0]) # 이 2차원 배열이니까 열의 개수는 첫 번째 행의 길이로 계산해.

    queue = deque ([0,0,1]) 

    visited = [[False]*m for _ in range(n)] # 이 리스트는 2차원 배열을 표현함. 
    visited[0][0] = True # 캐릭터가 (0, 0)에서 시작하므로, 이 위치를 이미 방문한 것으로 표시.  

    while queue:
        x,y,count = queue.popleft()

        if x == n-1 and y == m-1:
            return count 
        # 파이썬에서는 인덱스가 0부터 시작하기 때문에, 
        # 가장 끝 위치는 각각 행과 열의 인덱스가 **n-1과 m-1**이 되는 거야.

        for i in range(4): 
            nx = x+dx[i]
            ny = y+dy[i]

            if 0 <= nx < n and 0 <= ny < m and not visited[nx][ny] and maps[nx][ny] ==1 :
                visited[nx][ny] = True 
                queue.append((nx, ny, count +1))
    return -1 
```
