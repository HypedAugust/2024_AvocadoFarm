```python
# 문제2) 2차원 행렬에서 경로 탐색 (미로 찾기)

def dfs_maze(matrix, x,y, visited, path):
    n,m = len(matrix), len(matrix[0])

    if x == n -1 and y == m -1: # x,y는 인덱스라서 n-1임. 
        path.append((x,y))  
        return True 
    
    visited.add((x,y)) 
    path.append((x,y))

    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    for dx, dy in directions:
        nx,ny = x+dx, y+dy

        if 0<= nx < n and 0<= ny <m and matrix[nx][ny] ==1 and (nx,ny) not in visited:
            if dfs_maze(matrix, nx, ny, visited, path):
                return True 
    
    path.pop()
    return False 

# 미로 표현 (0은 벽, 1은 이동 가능한 경로)
maze = [
    [1, 0, 1, 1],
    [1, 1, 0, 1],
    [0, 1, 1, 1],
    [1, 0, 0, 1]
]

visited = set()
path = []

dfs_maze(maze, 0, 0, visited, path)
```

```python
# 문제1) 큐를 이용한 숫자 정렬 
# FIFO인점을 잊지말자. 
# 큐를 사용해 숫자들을 입력받고, 숫자들이 모두 큐에 들어간 후 순서대로 꺼내서 
# 정렬하여 출력해보자. 

from collections import deque 

def sort_numbers_with_queue(numbers):
    # 큐를 생성 
    queue = deque()

    # 모든 숫자를 큐에 넣기 
    for number in numbers:
        queue.append(number)

    # 큐에서 숫자를 꺼내 리스트에 저장 
    sorted_list = []
    while queue:
        sorted_list.append(queue.popleft())

    # 리스트를 정렬하고 반환 
    sorted_list.sort()
    return sorted_list 

numbers = [3, 1, 4, 1, 5, 9]
print(sort_numbers_with_queue(numbers))
```
