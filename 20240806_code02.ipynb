polynomials 다항식  
여러개의 term이 이루는 수학적 표현   
각 항은 constant와 variable로 이루어져 있다.   
변수의 지수는 0 또는 양의 정수이다. 

```python
# python으로 polynomial과 polynomial equation 공부하기 

# Evaluating Polynomials 
# 다항식 정의 
def polynomial(x):
    return 2*x**3 + 3*x**2 - 4*x + 5 

# x 값에 대해 다항식 계산하기 
result = polynomial(2)
print(f"The value of the polynomial at x = {x} is result")
```

```python
# solving polynomial equations 
import numpy as np 

# 다항식의 계수를 배열로 정의 
coefficients = [2,3,-4,5]

# 방정식의  solution 찾기 
roots = np.roots(coefficients)
print(f'The roots of the polynomial equation are {roots}')
```

```python
# plotting polynomials 

import numpy as np 
import matplotlib.pyplot as plt  

# 다항식 정의 
def polynomial(x):
    return 2*x**3 + 3*x**2 - 4*x + 5

# x값의 범위 설정 
x = np.linspace(-10,10,400)
y = polynomial(x)

# 그래프 그리기 
# 그래프 그리기
plt.plot(x, y)
plt.title("Graph of the Polynomial 2x^3 + 3x^2 - 4x + 5")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# 다항식을 정의 하는 함수 
def polynomial(x): 
    return 2*x**3 + 3*x**2 -4*x +5

# 여러 x값에 대해 다항식 계산하기 
x_values = [0,1,2,3,4,5]
results = [polynomial(x) for x in x_values]

for x, result in zip(x_values,results):
    print(f"The value of the polynomial at x={x} is {result}")
```
