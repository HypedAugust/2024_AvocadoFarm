````python
import random

# seed 
def generate_coupon_code(n):
    final_answers = []
    characters = "abcdefghijklmnopqrstuvwxyz01234567890ABCDEFGHIJKLMNOPQRSTUVWXYZ!@#$%^&*()?"
    random.seed(None)

    for i in range(0,n):
        answers = ''.join(random.sample(characters,6))
        print(answers)

print(generate_coupon_code(5))
````
