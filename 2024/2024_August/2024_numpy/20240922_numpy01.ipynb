```python
# 실습 예제 9 

# 10진수 100 <= x <= 150 사이에 존제 하는 정수를 randint로 무작위 추출 
# (3,10) 형태 행렬화 
# 행과 열 전치할 것. 

import numpy as np

# 내답
arr = np.random.randint(100,150, size = (3,10))
print(arr.T)

# 정답 
arr = np.random.randint(low=100, high= 150, size =(3,10))
print(np.transpose(arr))
```

```python
# 실습 예제 10 

# 10-20 사이 실수 무작위 10000 추출할 것 
# 100개 구간에 그래프로 시각화 하기 
# pyplot 모듈 쓰기 
# rand는 0,1 사이의 실수를 무작위 생성이다. 10-20은 안된다

import numpy as np
import matplotlib.pyplot as plt 


# 10~20 사이의 실수 10000개를 무작위로 추출
arr = np.random.uniform(10, 20, 10000) # uniform은 원하는 범위의 실수를 생성한다. 

# 100개 구간으로 히스토그램 그리기
plt.hist(arr, bins=100)
plt.show()
```

```python
# 다른 답 
randoms = 10 + np.random.rand(10000)*10 # 왜 10을 곱하는거야? sacling up을 위해서야! 
plt.hist(randoms,bins = 100)
plt.show()
```
