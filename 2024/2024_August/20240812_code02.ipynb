```python
# os.walk 
import os 
path = rf'C:\Users\qkim8\Desktop\work_kyu\43-1'

for dirpath, dirnames, filenames in os.walk(path):
    print(f'found directory: {dirpath}')
    for file_name in filenames:
        print(file_name)
```
