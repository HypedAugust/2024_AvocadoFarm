```python
# 문제2) 2차원 행렬에서 경로 탐색 (미로 찾기)

def dfs_maze(matrix, x,y, visited, path):
    n,m = len(matrix), len(matrix[0])

    if x == n -1 and y == m -1: # x,y는 인덱스라서 n-1임. 
        path.append((x,y))  
        return True 
    
    visited.add((x,y)) 
    path.append((x,y))

    directions = [(1, 0), (-1, 0), (0, 1), (0, -1)]

    for dx, dy in directions:
        nx,ny = x+dx, y+dy

        if 0<= nx < n and 0<= ny <m and matrix[nx][ny] ==1 and (nx,ny) not in visited:
            if dfs_maze(matrix, nx, ny, visited, path):
                return True 
    
    path.pop()
    return False 

# 미로 표현 (0은 벽, 1은 이동 가능한 경로)
maze = [
    [1, 0, 1, 1],
    [1, 1, 0, 1],
    [0, 1, 1, 1],
    [1, 0, 0, 1]
]

visited = set()
path = []

dfs_maze(maze, 0, 0, visited, path)
```

```python
# 문제1) 큐를 이용한 숫자 정렬 
# FIFO인점을 잊지말자. 
# 큐를 사용해 숫자들을 입력받고, 숫자들이 모두 큐에 들어간 후 순서대로 꺼내서 
# 정렬하여 출력해보자. 

from collections import deque 

def sort_numbers_with_queue(numbers):
    # 큐를 생성 
    queue = deque()

    # 모든 숫자를 큐에 넣기 
    for number in numbers:
        queue.append(number)

    # 큐에서 숫자를 꺼내 리스트에 저장 
    sorted_list = []
    while queue:
        sorted_list.append(queue.popleft())

    # 리스트를 정렬하고 반환 
    sorted_list.sort()
    return sorted_list 

numbers = [3, 1, 4, 1, 5, 9]
print(sort_numbers_with_queue(numbers))
```

```python
# 문제2) Linked List에서 특정 값 삭제
# 단순 연결 리스트에서 특정 값을 삭제하는 함수를 작성하세요.

class LinkedList:
    def __init__(self):
        self.head = None
    
    def append(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return 

        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def delete(self, value):
        current = self.head # 연결 리스트의 시작 지점인 'head'에서 시작

        # if head needs to be deleted
        if current and current.data == value: 
            self.head = current.next # 값이 맞으면 현재 옆으로 가버림. 
            return # 옆으로 가버렸으니까 지운거고 함수 종료

        prev = None # 일단 비워둠. 예전 값을 여따 넣을 거임 

        while current and current.data != value: # 현재 노드가 지우려는 값이 아니면
            prev = current # 계속 옆으로 감. 
            current = current.next  # 그리고 'current'를 다음 노드로 옮김 

        if current: # 'current'가 None이 아니면 (즉, 우리가 찾는 값을 가진 노드를 찾았다면)
            prev.next = current.next # 그 CURRENT 스킵하고 바로 옆에 값이렁 연결해버림 (삭제)

    def display(self):
        current = self.head
        while current: 
            print(current.data, end = ' ')
            current = current.next
        print()
```

```python
# 문제1) 이진 트리의 전위 순회 (Preorder Traversal)
# 이진 트리를 전위 순회하는 코드를 작성하세요.

class TreeNode:
    def __init__(self, data):
        self.data = data 
        self.left = None
        self.right = None
    

    def preorder_traversal(node):
        if node: # 현재 노드가 존재할 때만 그 아래 작업을 수행
            print(node.data, end= ' ') # 현재 노드 출력
            preorder_traversal(node.left) # 현재 노드의 왼쪽 자식을 전위 순회
            preorder_traversal(node.right)
            

# 테스트
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

# "1부터 시작(루트) → 2(왼쪽 자식) → 4(왼쪽 자식의 왼쪽 자식) → 5(왼쪽 자식의 오른쪽 자식) → 3(오른쪽 자식)"
preorder_traversal(root)  # 출력: 1 2 4 5 3
```

```python
# 이해하기 어려움 
# 문제2) 이진 트리의 높이 구하기
# 이진 트리의 높이를 계산하는 코드를 작성하세요.

def tree_height(node):
    if not node:
        return 0 
    
    left_height = tree_height(node.left)
    right_height = tree_height(node.right)

    return max(left_height, right_height) + 1 # 엣지가 아니고 노드 갯수 세야함. 
```

```python
# 문제1) 해시 테이블로 단어 개수 세기
# 문자열이 주어졌을 때, 
# 각 단어의 빈도를 세어 해시 테이블 형태로 반환하세요.

def word_count(s):
    word_dict = {}
    words = s.split() # split() 함수는 문자열을 **빈칸을 기준으로** 나누어 각 단어로 분리합니다.
                       # "hello world hello" -> ['hello', 'world', 'hello']
                       # 빈칸은 단어 사이의 구분자로 사용되며, 분리된 결과에는 포함되지 않아요.


    for word in words:
        if word in word_dict: 
            word_dict[word] += 1 
        else:
            word_dict[word] = 1 # # 만약 단어가 word_dict에 없다면, 새로 추가하고 개수를 1로 설정해요

    return word_dict 

# 테스트
s = "hello world hello"
print(word_count(s)) 

# 파이썬의 **딕셔너리(Dictionary)**가 해시 테이블의 한 형태인데요, 
# 딕셔너리는 키-값 쌍을 저장하고, 
# 키를 이용해 값을 빠르게 찾을 수 있는 자료 구조입니다.

# 해시는 데이터를 고정된 크기의 해시 값으로 변환하는 것을 의미하고, 
# 해시 테이블은 해시 값을 이용해서 빠르게 검색하고 저장하는 자료 구조입니다.
```

```python
# 문제2) 해시 테이블로 배열에서 중복 찾기
# 배열에서 중복된 값을 찾고, 
# 그 값들을 반환하는 함수를 작성하세요.

ddef find_duplicates(arr):
    elements_count = {}  # 각 요소의 개수를 세기 위한 딕셔너리
    duplicates = []  # 중복된 값을 저장할 리스트

    # 첫 번째 반복문: 배열의 각 요소를 순회하면서 중복 여부를 확인
    for elem in arr:
        # 딕셔너리로 중복되는 수를 세는 작업
        if elem in elements_count:
            elements_count[elem] += 1  # 이미 있는 요소면 개수를 1 증가
        else:
            elements_count[elem] = 1  # 처음 등장한 요소는 1로 설정

    # 두 번째 반복문: 딕셔너리에서 개수가 2 이상인 요소를 찾기
    for key, value in elements_count.items():
        # 개수가 2 이상이면 중복된 값이므로
        if value > 1:
            duplicates.append(key)  # 중복된 값을 duplicates 리스트에 추가

    return duplicates
```
